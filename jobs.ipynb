```python
from splinter import Browser
from bs4 import BeautifulSoup
import requests
import pymongo
executable_path = {'executable_path': '/Users/raghi/Downloads/chromedriver'}
browser = Browser('chrome', **executable_path, headless=True)
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
db = client.job_data
job_collection = db.job_collection
states_to_collect = [
    {"state": "NJ", "state_name": "New Jersey"},
    {"state": "PA", "state_name": "Pennsylvania"},
    {"state": "NC", "state_name": "North Carolina"},
    {"state": "MA", "state_name": "Massachusetts"},
    {"state": "CT", "state_name": "Connecticut"},
    {"state": "DE", "state_name": "Delaware"},
    {"state": "MD", "state_name": "Maryland"},
    {"state": "NY", "state_name": "New York"},
    {"state": "MN", "state_name": "Minnesota"},
    {"state": "FL", "state_name": "Florida"},
    
]
search_query = "data+scientist"
for st in states_to_collect:
    state = st["state"]
    state_name = st["state_name"]
    url = "https://www.indeed.com/jobs?q={}&l={}&radius=0&sort=date&limit=50&fromage=15&start={}"
    browser.visit(url.format(search_query, state, 0))
    s = BeautifulSoup(browser.html, "html.parser")
    searchCount = s.find('div', id="searchCount")
    pages = int(searchCount.text.strip().split()[3])// 50 + 1
    c = 0
    data = []
    for i in range(pages):
        browser.visit(url.format(search_query, state, 50))
        s = BeautifulSoup(browser.html, "html.parser")
        items = s.find_all('div', class_='jobsearch-SerpJobCard')
        for item in items:
            try:
                job = item.find('h2', class_="jobtitle").a.text
                location = item.find('span', class_="location").text
                c += 1
                print(c, job, location)
                data.append({"title": job, "location": location, "state": state_name, "search_query": search_query})
            except:
                pass
    # insert to mongo collection.    
    job_collection.insert_many(data)

```
